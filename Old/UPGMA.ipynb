```python
from scipy.cluster.hierarchy import dendrogram, linkage, to_tree
from scipy.spatial.distance import squareform
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

# names of the leaves
species = ['Aadvark', 'Alligator M', 'Alligator S', 'Anolis', 'Chelonia', 'Chrysemys', 'Croco', 'Danio', 'Devil', 'Dog', 'Fugu', 'Gecko', 'Gorilla', 'Human', 'KCobra', 'Koala', 'Monodelphis', 'Mouse', 'Orca', 'Ostrich', 'Papio', 'Pelodiscus', 'Pigeon', 'Platypus', 'Pogona', 'Python', 'Thamnophis', 'Xenopus', 'Xlaevis', 'Xtropicalis']

filepath = '..\Data\Results\Metric\metric_50_m_n.xlsx'
head, tail = os.path.split(filepath)
filename = tail.split('.')[0]

matrix = np.nan_to_num(pd.read_excel(filepath, index_col=0).to_numpy(), nan=0)
matrix = squareform(matrix)

# calculate linkage matrix
Z = linkage(matrix, method='average')

# plot dendrogram
plt.figure(figsize=(25, 10))
plt.title('Phylogenetic tree for ' + filename, fontsize=20)
dendrogram(Z, labels=species, leaf_rotation=90, leaf_font_size=14)
plt.show()
```
